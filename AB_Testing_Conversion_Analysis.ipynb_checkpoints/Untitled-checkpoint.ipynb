```python
import pandas as pd
import numpy as np
```

```python
# Step 1: Data Loading & Cleaning (Sanity Check)
```

```python
df = pd.read_csv('AB_Test_Results.csv')
print("原始列名为:", df.columns.tolist())
```

```python
df.columns = df.columns.str.strip().str.upper()
```

```python
print("原始数据量:", df.shape)
```

```python
user_counts = df.groupby('USER_ID')['VARIANT_NAME'].nunique()
users_in_both = user_counts[user_counts > 1].index
print(f"同时在两组的异常用户数: {len(users_in_both)}")
```

```python
df_clean = df[~df['USER_ID'].isin(users_in_both)].copy()
```

```python
df_clean['CONVERTED'] = (df_clean['REVENUE'] > 0).astype(int)
print("清洗后数据量:", df_clean.shape)
```

```python
# Step 2: Exploratory Data Analysis
```

```python
summary = df_clean.groupby('VARIANT_NAME').agg(
    users=('USER_ID', 'nunique'),
    conversions=('CONVERTED', 'sum'),
    total_revenue=('REVENUE', 'sum')
)
```

```python
summary['conversion_rate'] = summary['conversions'] / summary['users']
summary['ARPU'] = summary['total_revenue'] / summary['users']
```

```python
summary_display = summary.copy()
summary_display['conversion_rate'] = summary_display['conversion_rate'].map('{:.2%}'.format)
summary_display['ARPU'] = summary_display['ARPU'].map('${:.4f}'.format)
print("---------- 实验基础数据表现 ----------")
print(summary_display)
print("\n")
```

```python
# Step 3: Statistical Hypothesis Testing
```

```python
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind
```

```python
control_conv = summary.loc['control', 'conversions']
variant_conv = summary.loc['variant', 'conversions']
control_users = summary.loc['control', 'users']
variant_users = summary.loc['variant', 'users']
```

```python
z_stat, pval_z = proportions_ztest([variant_conv, control_conv], [variant_users, control_users])
print("---------- 转化率检验结果 ----------")
print(f"P-value: {pval_z:.4f}")
if pval_z < 0.05:
    print("结论: 拒绝原假设，新老版本转化率有统计学上的显著差异！")
else:
    print("结论: 无法拒绝原假设，新老版本转化率没有显著差异。")
print("\n")
```

```python
control_rev = df_clean[df_clean['VARIANT_NAME'] == 'control']['REVENUE']
variant_rev = df_clean[df_clean['VARIANT_NAME'] == 'variant']['REVENUE']

t_stat, pval_t = ttest_ind(variant_rev, control_rev, equal_var=False)
print("---------- 人均创收(ARPU)检验结果 ----------")
print(f"P-value: {pval_t:.4f}")
if pval_t < 0.05:
    print("结论: 拒绝原假设，人均创收有显著差异！")
else:
    print("结论: 无法拒绝原假设，人均创收没有显著差异。")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
```

```python
ax1 = sns.barplot(x='VARIANT_NAME', y='CONVERTED', data=df_clean, 
                  ci=95, capsize=0.1, palette=['#4A90E2', '#50E3C2'])

plt.title('Conversion Rate by Group (95% Confidence Interval)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Experiment Group', fontsize=12)
plt.ylabel('Conversion Rate (CVR)', fontsize=12)
```

```python
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.2%}'.format(y)))
```

```python
for p in ax1.patches:
    ax1.annotate(f"{p.get_height():.2%}", 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='bottom', xytext=(0, 10), textcoords='offset points', 
                 fontsize=11, fontweight='bold')

plt.tight_layout()
```

```python
plt.savefig('cvr_comparison.png', dpi=300)
plt.show()
```
